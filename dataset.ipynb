```python
import random, re
from pythainlp.tokenize import word_tokenize

RE_EN = re.compile(r"[A-Za-z]{2,}|[ /+-]")

def typo_sentence(sentence, max_tries=10):
    tokens = word_tokenize(str(sentence), engine="newmm")
    thai_tokens = [t for t in tokens if not RE_EN.search(t)]
    if not thai_tokens:
        return sentence

    token = random.choice(thai_tokens)
    chars = list(token)
    for i, ch in enumerate(chars):
        if ch in typo_dict:
            chars[i] = random.choice(typo_dict[ch])
    new_token = "".join(chars)

    idx = tokens.index(token)
    tokens[idx] = new_token
    return "".join(tokens)
```

```python
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm.auto import tqdm

results = [None] * len(df)
errors = []

with ProcessPoolExecutor(max_workers=30) as exe:
    futures = {
        exe.submit(typo_sentence, x): i
        for i, x in enumerate(df["inputs"])
    }
    for f in tqdm(as_completed(futures), total=len(futures), desc="Typing typos"):
        idx = futures[f]
        try:
            results[idx] = f.result()
        except Exception as e:
            errors.append((idx, e))

df["typo_message"] = results
```

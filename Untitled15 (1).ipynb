```python
import re
from itertools import chain

from pythainlp.tokenize import word_tokenize
from pythainlp.corpus.common import thai_words

THAI_WORDS = set(thai_words())

try:
    import pyenchant as enchant

    EN_DICT = enchant.Dict("en_US")
    def is_english_word(word):
        return EN_DICT.check(word)
    EN_WORDS = None
except ImportError:
    from nltk.corpus import words as nltk_words_list

    EN_WORDS = set(nltk_words_list.words())
    def is_english_word(word):
        return word.lower() in EN_WORDS

ALPHABETS = {
    'thai': list({c for w in THAI_WORDS for c in w}),
    'eng': list('abcdefghijklmnopqrstuvwxyz')
}

ENG_TO_THAI = {
    '1': 'ๅ',    '2': '/',    '3': '-',    '4': 'ภ',   '5': 'ถ',   '6': 'ุ',   '7': 'ึ',
    '8': 'ค',    '9': 'ต',    '0': 'จ',    '-': 'ข',   '=': 'ช',   'q': 'ๆ',   'w': 'ไ',
    'e': 'ำ',    'r': 'พ',    't': 'ะ',    'y': 'ั',   'u': 'ี',   'i': 'ร',   'o': 'น',
    'p': 'ย',    '[': 'บ',    ']': 'ล',    '\\': 'ฃ',  'a': 'ฟ',   's': 'ห',   'd': 'ก',
    'f': 'ด',    'g': 'เ',    'h': '้',    'j': '่',   'k': 'า',   'l': 'ส',   ';': 'ว',
    "'": 'ง',    'z': 'ผ',    'x': 'ป',    'c': 'แ',   'v': 'อ',   'b': 'ิ',   'n': 'ื',
    'm': 'ท',    ',': 'ม',    '.': 'ใ',    '/': 'ฝ',   '!': '+',   '@': '๑',  '#': '๒',
    '$': '๓',    '%': '๔',    '^': 'ู',    '&': '฿',   '*': '๕',   '(': '๖',   ')': '๗',
    '_': '๘',    '+': '๙',    'Q': '๐',    'W': '"',   'E': 'ฎ',   'R': 'ฑ',   'T': 'ธ',
    'Y': 'ํ',    'U': '๊',    'I': 'ณ',    'O': 'ฯ',   'P': 'ญ',   '{': 'ฐ',   '}': ',',
    '|': 'ฅ',    'A': 'ฤ',    'S': 'ฆ',    'D': 'ฏ',   'F': 'โ',   'G': 'ฌ',   'H': '็',
    'J': '๋',    'K': 'ษ',    'L': 'ศ',    ':': 'ซ',   '"': '.',   'Z': '(',   'X': ')',
    'C': 'ฺ',    'V': 'ฮ',    'B': 'ฺ',    'N': '์',   'M': '?',   '<': 'ฒ',   '>': 'ฬ',
    '?': '"'
}

RE_THAI = re.compile(r'[ก-๙]')
RE_ENG = re.compile(r'^[A-Za-z]+$')

def edits1(word: str, alphabet: list) -> set:
    """Return all strings one edit away from `word`."""
    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in alphabet]
    inserts    = [L + c + R               for L, R in splits for c in alphabet]
    return set(deletes + transposes + replaces + inserts)


def edits2(word: str, alphabet: list) -> set:
    """Return all strings two edits away from `word`."""
    return set(e2 for e1 in edits1(word, alphabet) for e2 in edits1(e1, alphabet))


def known(words: set, dictionary) -> set:
    """Filter `words` returning only those present in `dictionary`."""
    if hasattr(dictionary, 'check'):
        return set(w for w in words if dictionary.check(w))
    else:
        return set(w for w in words if w.lower() in dictionary)


def candidates(word: str, dictionary, alphabet: list) -> set:
    """Generate possible spelling corrections for `word`."""
    c1 = known({word}, dictionary)
    if c1:
        return c1
    c2 = known(edits1(word, alphabet), dictionary)
    if c2:
        return c2
    c3 = known(edits2(word, alphabet), dictionary)
    if c3:
        return c3
    return {word}

def correct_word(word: str, dictionary, alphabet: list) -> str:
    """Return the best candidate correction for `word`."""
    cands = candidates(word, dictionary, alphabet)
    return max(cands, key=len)

def correct_token(tok: str) -> str:

    if RE_ENG.match(tok):
        return tok

    mapped = ''.join(ENG_TO_THAI.get(ch, ch) for ch in tok)
    if mapped != tok and RE_THAI.search(mapped):
        return correct_word(mapped, THAI_WORDS, ALPHABETS['thai'])

    if RE_THAI.search(tok):
        return correct_word(tok, THAI_WORDS, ALPHABETS['thai'])

    return tok


def thai_english_speller(text: str) -> str:
    """Tokenize, correct each token, and rejoin without adding extra spaces."""
    tokens = word_tokenize(text, engine='newmm')
    corrected = [correct_token(t) for t in tokens]
    return ''.join(corrected)

if __name__ == '__main__':
    examples = [
        "เราควร backup -hv,^]]' external Hard Drivem6d,gzv,xV",
        "ช่วยส่งไฟล์Excel,k.shz,sojvpwfhws,8iy[",
        "chinhenk",
        "มีการยานเยอะมากทําไมไหวเเล้ว",
        "หวัดดีเป็นไงบ้างอะkenmkw,w,Fmo,kg]p",
        "ใหม่! กันแดดอัจฉริยะ UV Adapt Hya Water Sunscreen SPF50! PA!!!! xdxhv'zb;]he]7d57' 2 9jv!"
    ]
    for s in examples:
        print(f"Input : {s}")
        print(f"Output: {thai_english_speller(s)}")
        print('-' * 40)
```

```python
import pandas as pd
from Levenshtein import distance as levenshtein_distance
from nltk.translate.bleu_score import corpus_bleu
from tqdm.auto import tqdm  # เพิ่ม tqdm

# โหลดข้อมูล
test_df = pd.read_csv("realworld.csv")
refs = test_df["sentence"].tolist()

# สร้างพยากรณ์ (preds) พร้อม progress bar
preds = [
    thai_english_speller(t)
    for t in tqdm(test_df["eng_sentence"], desc="ตรวจแก้คำผิด")
]

# Character-Level Accuracy
total_chars = sum(len(r) for r in refs)
correct_chars = sum(
    sum(1 for a, b in zip(r, p) if a == b)
    for r, p in tqdm(zip(refs, preds), total=len(refs), desc="นับตัวอักษรถูกต้อง")
)
char_acc = correct_chars / total_chars

# Average Levenshtein Distance
avg_edit = (
    sum(
        levenshtein_distance(r, p)
        for r, p in tqdm(zip(refs, preds), total=len(refs), desc="คำนวณ Levenshtein")
    )
    / len(refs)
)

# Exact Match Accuracy
exact_match = (
    sum(1 for r, p in tqdm(zip(refs, preds), total=len(refs), desc="ตรวจ Exact Match") if r == p)
    / len(refs)
)

# BLEU Score (ไม่จำเป็นต้องใช้ tqdm เพราะเร็ว)
list_of_refs = [[r.split()] for r in refs]
hypotheses  = [p.split()    for p in preds]
bleu_score  = corpus_bleu(list_of_refs, hypotheses)

# แสดงผล
print(f"Character-Level Accuracy : {char_acc:.4f}")
print(f"Average Edit Distance    : {avg_edit:.2f}")
print(f"Exact Match Accuracy     : {exact_match:.4f}")
print(f"BLEU Score               : {bleu_score:.4f}")
```
